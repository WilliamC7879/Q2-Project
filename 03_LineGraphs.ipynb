## Example Graphs

### `plot_lines(names, date_range=None, nametype='Site Name', window=10)`
These graphs plot each site. If given a county, every site within that county is plotted indiviudally.

```python
plot_lines('San Francisco', [2017,2017])
```

```python
plot_lines(['San Francisco', 'San Jose', 'Berkeley'], [2015,2019])
plot_lines('Marin', [2015,2019], 'COUNTY')
```

```python
plot_lines(['Paradise', 'San Francisco'], [2017,2019], window=10)

plt.figure()
plot_lines(['Livermore', 'Fresno', 'San Jose', 'Pasadena',
            'New York'], [2010,2018], window=30)
```

```python
plot_lines('Marin', nametype='COUNTY')

plt.figure()
plot_lines(['Orange', 'Butte'], [2015,2019], 'COUNTY')
```

### `plot_mean(counties, date_range=None, window=10)`
These graphs plot each county as one line, by averaging the data from each site in that county per day.

```python
plot_mean(['Marin', 'Santa Cruz'])

plt.figure()
plot_mean('Los Angeles', [2018,2019], window=3)
```
