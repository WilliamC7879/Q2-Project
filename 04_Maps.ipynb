**Installation:** (it's quite a ride...)
1. [Make new environment](https://github.com/geopandas/geopandas/issues/237#issuecomment-415351478):
  - `conda create -n gpd -c conda-forge python=3.6 geopandas`
2. Activate environment:
  - `source activate gpd`
3. [Install ipython in environment](https://stackoverflow.com/questions/33960051/unable-to-import-a-module-from-python-notebook-in-jupyter):
  - `conda install notebook ipykernel`
  - `ipython kernel install --user`
4. Open jupyter -- make sure you're starting jupyter in this environment, or else geopandas won't work.
  - `jupyter notebook`

### `def map_counties(dates, save=False)`

These are choropleth maps, which show intensity of a variable (in this case, AQI) through color. Given a date or list of dates, `map_counties` will show a number of choropleth maps with a colorbar.

```python
map_counties(['11/13/17'])
```

```python
map_counties([dt(2017,9,day) for day in range(1,6,2)])
```

```python
map_counties([dt(yr,11,15) for yr in range(2015,2019)])
```

```python
map_counties([dt(2018,11,5), dt(2018,11,10), dt(2018,11,15), dt(2018,11,20), dt(2018,11,25)])
```
